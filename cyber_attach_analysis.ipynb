```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv("all_recs.csv", encoding= 'unicode_escape')
```

```python
year_order= df['Year'].value_counts().index
#value count gets the unique occurnce of the individual years as well as arranges them in order of frequency
location_order = df['location'].value_counts().index
motive_order = df['Motive'].value_counts().index
motive_order = df['industry'].value_counts().index
```

```python
import matplotlib.pyplot as plt
import seaborn as sb

year_order = df['Year'].value_counts().index

plt.figure(figsize=(20, 8))
plt.title("Attacks frequency by Year")

sb.countplot(
    data = df, 
    y = 'Year', 
    color = sb.color_palette()[2], 
    order = year_order
)


plt.yticks(rotation = 45)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sb

motive_order = df['Motive'].value_counts().index

plt.figure(figsize=(20, 8))
plt.title("Attacks frequency by Motive")

sb.countplot(
    data = df,
    y = 'Motive',
    color = sb.color_palette()[2],
    order = motive_order
)

plt.yticks(rotation = 45)
plt.show()
```
