# Cyber Attack Analysis
Analysis of attack frequency by Motive and Location

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12,6)
```

```python
# Load data
df = pd.read_csv('all_recs.xlsx_-_Sheet1.csv', encoding='unicode_escape')

# Fill missing values for plotting
df['Motive'] = df['Motive'].fillna('Missing/NaN')
df['Location'] = df['Location'].fillna('Missing/NaN')

df.head()
```

```python
# Dataset overview
print(f'Shape: {df.shape[0]:,} rows × {df.shape[1]} columns')
print('\nColumn info:')
df.info()

print('\nMissing values per column:')
print(df.isna().sum())
```

## Attack Frequency by Motive

```python
# Motive plot - highest to lowest
motive_counts = df['Motive'].value_counts().sort_values(ascending=True)

plt.figure(figsize=(12,6))
sns.barplot(x=motive_counts.values, y=motive_counts.index, color='#4C72B0')
plt.title('Attack Frequency by Motive (Highest → Lowest)', fontsize=16, fontweight='bold')
plt.xlabel('Number of Attacks')
plt.ylabel('Motive')

# Add value labels
for i, v in enumerate(motive_counts.values):
    plt.text(v + 50, i, f'{v:,}', va='center', fontsize=9)

plt.tight_layout()
plt.show()
```

## Attack Frequency by Location

```python
# Location plot - top 25
top_n = 25
loc_counts = df['Location'].value_counts().head(top_n).sort_values(ascending=True)

plt.figure(figsize=(12,9))
sns.barplot(x=loc_counts.values, y=loc_counts.index, color='#DD6B20')
plt.title(f'Top {top_n} Attack Frequency by Location (Highest → Lowest)', fontsize=16, fontweight='bold')
plt.xlabel('Number of Attacks')
plt.ylabel('Location')

# Add value labels
for i, v in enumerate(loc_counts.values):
    plt.text(v + 30, i, f'{v:,}', va='center', fontsize=8)

plt.tight_layout()
plt.show()
```

## Conclusion

The Motive graph shows the most common reasons behind cyber attacks.
The Location graph shows the locations with the highest number of cyber incidents.

Both graphs were generated using Seaborn count plots and sorted from highest to lowest frequency.
